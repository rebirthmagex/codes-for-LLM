#**4. Características Avançadas do Docling**

##4.1 Configurando um pipeline de análise

```python
IMAGE_RESOLUTION_SCALE = 2.0

input_doc_path = Path("sample_doc.pdf")

pipeline_options = PdfPipelineOptions()
pipeline_options.images_scale = IMAGE_RESOLUTION_SCALE
pipeline_options.generate_page_images = True
pipeline_options.generate_picture_images = True

doc_converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
    }
)

start_time = time.time()
conv_res = doc_converter.convert(input_doc_path)
end_time = time.time() - start_time
print(f"Tempo total necessário para análise: {end_time:.2f} segundos.")
```

##4.2 Extraindo Metadados do Documento

```python
print("Document Name:", conv_res.document.origin.filename)
print("Document Type:", conv_res.document.origin.mimetype)
print("Number of Pages:", len(conv_res.document.pages.keys()))
print("Number of text elements:", len(conv_res.document.dict()["texts"]))
print("Number of tables:", len(conv_res.document.dict()["tables"]))
print("Number of images:", len(conv_res.document.dict()["pictures"]))
```

###4.2.1 Iterando pelos elementos compostos por texto

```python
dict_list = []

texts = conv_res.document.dict()["texts"]

for t in texts:
    ref = t["self_ref"]
    page = t["prov"][0]["page_no"]

    text_content = t["text"]

    dict_list.append({"text_reference": ref, "page": page, "text_content(first 500 chars)": text_content[:500]})

df = pd.DataFrame(dict_list)
display(df)
```

###4.2.2 Iterando pelos elementos compostos por tabela

```python
dict_list = []

tables = conv_res.document.dict()["tables"]

for t in tables:
    ref = t["self_ref"]
    page = t["prov"][0]["page_no"]

    table_captions = t["captions"]

    if (len(table_captions) and 'cref' in table_captions[0].keys()):
        table_caption = table_captions[0]["cref"]

    else:
        table_caption = "No Caption"

    table_data = str(t["data"])[:100]

    dict_list.append({"table_reference": ref, "page": page, "table_captions": table_caption, "table_data(first 500 chars)": table_data})

df = pd.DataFrame(dict_list)
display(df)
```

###4.2.3 Iterando pelos elementos compostos por imagem

```python
dict_list = []

images = conv_res.document.dict()["pictures"]

for i in images:
    ref = i["self_ref"]
    page = i["prov"][0]["page_no"]

    image_captions = i["captions"]

    if (len(image_captions) and 'cref' in image_captions[0].keys()):
        image_caption = image_captions[0]["cref"]

    else:
        image_caption = "No Caption"

    image_data = str(i["image"])[:100]

    dict_list.append({"image_reference": ref, "page": page, "image_captions": image_caption, "image_data(first 500 chars)": image_data})

df = pd.DataFrame(dict_list)
display(df)
```

##4.3 Iterando pelas Páginas do Documento  

```python
# Exibindo as páginas do documento
def display_images(images, images_per_row=5, figsize=(15, 8)):

    # Criando o dicionário ou lista de entrada
    if isinstance(images, dict):
        images = list(images.values())

    # Calculando o número de linhas necessárias
    num_rows = math.ceil(len(images) / images_per_row)

    # Criando os subplots
    fig, axes = plt.subplots(num_rows, images_per_row, figsize=figsize)
    axes = axes.flatten()

    # Plotando as imagens
    for ax, image in zip(axes, images):
        ax.imshow(image)
        ax.axis('off')

    # Desativando os eixos
    for ax in axes[len(images):]:
        ax.axis('off')

    # Ajustando o layout e o display
    plt.tight_layout()
    plt.show()

page_images = {page_no: page.image.pil_image for page_no, page in conv_res.document.pages.items()}
display_images(page_images, images_per_row=5, figsize=(15, 8))
```

###4.3.1 Gerando imagens das páginas do documento

```python
# Salvando as páginas como imagens
dir_path = "./pages"
os.makedirs(dir_path, exist_ok=True)

for page_no, page in conv_res.document.pages.items():
    page_filename = os.path.join(dir_path, f"page_{page_no}.png")

    with open(page_filename, "wb") as fp:
        page.image.pil_image.save(fp, format="PNG")
```

##4.4 Extraindo todas as imagens do documento

```python
# Extraindo todas as imagens do documento
dir_path = "./images"
os.makedirs(dir_path, exist_ok=True)

images_list = []
image_number = 1

for element, _level in conv_res.document.iterate_items():
    if isinstance(element, PictureItem):
        # Salvando tabelas como imagens
        element_image_filename = os.path.join(dir_path, str(image_number) + ".png")

        with open(element_image_filename, "wb") as fp:
            image = element.get_image(conv_res.document)
            image.save(fp, "PNG")
            images_list.append(image)

        image_number += 1
```

```python
display_images(images_list, images_per_row=5, figsize=(15, 8))
```

##4.5 Extraindo todas as tabelas do documento

```python
# Extraindo todas as tabelas do documento
dir_path = "./tables"
dir_paths = ["./tables/images", "./tables/CSVs", "./tables/HTMLs"]
[os.makedirs(path, exist_ok=True) for path in dir_paths]

table_list = []
table_number = 1

for element, _level in conv_res.document.iterate_items():
    if isinstance(element, TableItem):
        # Salvando tabelas como imagens
        element_table_filename = os.path.join(dir_path, "images", str(table_number) + ".png")

        with open(element_table_filename, "wb") as fp:
            table_image = element.get_image(conv_res.document)
            table_image.save(fp, "PNG")
            table_list.append(table_image)

        # Salvando tabelas como arquivos CSV
        table_df: pd.DataFrame = element.export_to_dataframe()
        element_table_filename = os.path.join(dir_path, "CSVs", str(table_number) + ".csv")
        table_df.to_csv(element_table_filename)

        # Salvando tabelas como arquivos HTML
        table_html = element.export_to_html()
        element_table_filename = os.path.join(dir_path, "HTMLs", str(table_number) + ".html")

        with open(element_table_filename, "w") as fp:
            fp.write(table_html)

        table_number += 1
```

```python
display_images(table_list, images_per_row=5, figsize=(15, 8))
```

##5.2 Exemplo de pipeline habilitado para OCR

```python
def OCR_parsing(doc_path):
    input_doc = Path(doc_path)
    pipeline_options = PdfPipelineOptions()
    pipeline_options.do_ocr = True
    pipeline_options.do_table_structure = True
    pipeline_options.table_structure_options.do_cell_matching = True
    ocr_options = EasyOcrOptions(force_full_page_ocr=True)
    pipeline_options.ocr_options = ocr_options
    pipeline_options.generate_page_images = True

    start_time = time.time()

    converter = DocumentConverter(
        format_options={
            InputFormat.PDF: PdfFormatOption(
                pipeline_options=pipeline_options
            )
        }
    )

    result = converter.convert(input_doc)
    end_time = time.time() - start_time

    print(f"Tempo total necessário para análise: {end_time:.2f} segundos.")

    return result
```
