## Executando o Ollama
---

Para utilizar o Ollama, é necessário que ele seja executado como um serviço em segundo plano, paralelamente aos seus scripts.

Como o Jupyter Notebooks foi projetado para executar blocos de código em sequência, isso dificulta a execução simultânea de dois blocos. Como solução alternativa, criaremos um serviço utilizando o módulo subprocess em Python, garantindo que nenhum bloco de código seja bloqueado durante a execução.

O serviço pode ser iniciado com o seguinte comando: `ollama serve`.

Além disso, a função `time.sleep(5)` adiciona um pequeno atraso para garantir que o serviço Ollama esteja completamente ativo antes de iniciar o download do modelo.

````python
import threading
import subprocess
import time

def run_ollama_serve():
  subprocess.Popen(["ollama", "serve"])

thread = threading.Thread(target=run_ollama_serve)
thread.start()
time.sleep(5)
````

````python
from neo4j import GraphDatabase

# Criar driver do Neo4j
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))

# Função para deletar o vectorstore
def delete_vectorstore():
    with driver.session() as session:
        # Deletar todos os nós de vetores (Chunks)
        session.run("MATCH (n:Chunk) DETACH DELETE n;")
        print("✅ Todos os vetores foram deletados.")

        # Deletar índice vetorial (caso exista)
        session.run("DROP INDEX question_cache IF EXISTS;")
        print("✅ Índice vetorial deletado.")

# Chamar função
delete_vectorstore()

# Fechar conexão com Neo4j
driver.close()
````

````python
# rag/index.js

import os
from langchain_community.graphs import Neo4jGraph
from langchain_community.vectorstores import Neo4jVectorStore
from langchain_ollama import ChatOllama, OllamaEmbeddings
from pathlib import Path

# ✅ Configuração do Neo4j
config = {
    "url": os.getenv("NEO4J_URI"),
    "username": os.getenv("NEO4J_USER"),
    "password": os.getenv("NEO4J_PASSWORD"),
    "text_node_properties": ["text"],
    "index_name": "javascript_index",
    "keyword_index_name": "javascript_keywords",
    "search_type": "vector",
    "node_label": "Chunk",
    "text_node_property": "text",
    "embedding_node_property": "embedding",
}

# ✅ Inicializar Modelo de Linguagem
model = ChatOllama(
    temperature=0,
    max_retries=2,
    model=os.getenv("NLP_MODEL"),
    base_url=os.getenv("OLLAMA_BASE_URL"),
)

# ✅ Inicializar Modelo de Embeddings
ollama_embeddings = OllamaEmbeddings(
    model="nomic-embed-text",
    base_url=os.getenv("OLLAMA_BASE_URL"),
)

# ✅ Carregar e Processar a Documentação de JavaScript
with open("rag/data/javascript.txt", "r", encoding="utf-8") as file:
    documents = [
        {"page_content": sentence.strip(), "metadata": {}}
        for sentence in file.read().split(".")
        if len(sentence.strip()) > 10  # Evitar documentos muito curtos ou vazios
    ]

# ✅ Inicializar VectorStore no Neo4j
neo4j_vector_index = Neo4jVectorStore.from_existing_graph(ollama_embeddings, config)

# ✅ Função para Adicionar Documentos ao VectorStore (se não existirem)
async def add_document_if_not_exists(doc):
    search_results = await neo4j_vector_index.similarity_search(doc["page_content"], 2)

    if search_results and search_results[0].page_content == f"\ntext: {doc['page_content']}":
        print(f"🚫 Documento duplicado: \"{doc['page_content']}\"")
    else:
        print(f"✅ Adicionando novo documento: \"{doc['page_content']}\"")
        await neo4j_vector_index.add_documents([doc])

# ✅ Adicionar Documentos ao Vector Store (se não existirem)
for doc in documents:
    await add_document_if_not_exists(doc)

# ✅ Função para Responder Perguntas com Base no Contexto Armazenado
async def answer_question(question):
    print(f"🔍 Processando pergunta: \"{question}\"")

    # Buscar trechos mais relevantes no Neo4j
    results = await neo4j_vector_index.similarity_search_with_score(question, 1)
    relevant_chunks = [result[0].page_content.replace("text: ", "") for result in results if result]

    if not relevant_chunks:
        print("⚠️ Nenhum contexto relevante encontrado.")
        return "Desculpe, não encontrei informações suficientes para responder."

    # Construir o prompt para o modelo de IA
    context = "\n".join(relevant_chunks)
    prompt = f"""
    Responda à pergunta de forma concisa e natural com base no seguinte contexto:
    Não utilize informações fora do contexto fornecido.

    Contexto:
    {context}

    Pergunta: {question}

    Forneça uma resposta direta e informativa:
    """

    # Gerar resposta usando o modelo de IA
    response = await model.invoke(prompt)
    return response

# ✅ Perguntas de Exemplo
questions = [
    "Is JavaScript an object-oriented programming language?",
    "Is JavaScript an interpreted language?",
]

for question in questions:
    answer = await answer_question(question)
    print(f"📝 Resposta: {answer}")
````

````python
from neo4j import GraphDatabase
from faker import Faker
import json
import os

# Configuração do Neo4j
NEO4J_URI = userdata.get('NEO4J_URI')
NEO4J_USER = "neo4j"
NEO4J_PASSWORD = userdata.get('NEO4J_PASSWORD')

# Criar driver do Neo4j
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))

# Instanciar Faker para gerar dados fictícios
faker = Faker()

# Carregar cursos a partir de um arquivo JSON
with open("seed/data/courses.json", "r", encoding="utf-8") as file:
    courses = json.load(file)

# Criar alunos fictícios
students = [
    {
        "id": faker.uuid4(),
        "name": faker.name(),
        "email": faker.email(),
        "phone": faker.phone_number(),
    }
    for _ in range(20)
]

# Criar registros de vendas
sales_records = [
    {
        "studentId": student["id"],
        "courseId": faker.random_element(courses)["name"],
        "status": faker.random_element(["paid", "refunded"]),
        "paymentMethod": faker.random_element(["pix", "credit_card"]),
        "paymentDate": faker.past_date().isoformat(),
        "amount": round(faker.pyfloat(min_value=0, max_value=2000, right_digits=2), 2),
    }
    for student in students
    for _ in range(faker.random_int(min=1, max=10))
]

# Criar registros de progresso para vendas pagas
progress_records = [
    {
        "studentId": sale["studentId"],
        "courseId": sale["courseId"],
        "progress": faker.random_int(min=0, max=100),
    }
    for sale in sales_records
    if sale["status"] == "paid"
]

# Função para inserir dados no Neo4j
def insert_data():
    with driver.session() as session:
        # Limpar banco de dados (opcional)
        # session.run("MATCH (n) DETACH DELETE n")
        # print("🧹 Database cleared!")

        session.run(
            """
            UNWIND $batch AS row
            MERGE (c:Course {name: row.name})
            ON CREATE SET c.url = row.url
            """,
            batch=courses
        )
        print("✅ Courses Inserted!")

        session.run(
            """
            UNWIND $batch AS row
            MERGE (s:Student {id: row.id})
            ON CREATE SET s.name = row.name, s.email = row.email, s.phone = row.phone
            """,
            batch=students
        )
        print("✅ Students Inserted!")

        session.run(
            """
            UNWIND $batch AS row
            MATCH (s:Student {id: row.studentId}), (c:Course {name: row.courseId})
            MERGE (s)-[p:PURCHASED]->(c)
            ON CREATE SET p.status = row.status, p.paymentMethod = row.paymentMethod,
                          p.paymentDate = row.paymentDate, p.amount = row.amount
            ON MATCH SET p.status = row.status, p.paymentMethod = row.paymentMethod,
                         p.paymentDate = row.paymentDate, p.amount = row.amount
            """,
            batch=sales_records
        )
        print("✅ Sales Inserted!")

        session.run(
            """
            UNWIND $batch AS row
            MATCH (s:Student {id: row.studentId})-[:PURCHASED {status: "paid"}]->(c:Course {name: row.courseId})
            MERGE (s)-[p:PROGRESS]->(c)
            ON CREATE SET p.progress = row.progress
            ON MATCH SET p.progress = row.progress
            """,
            batch=progress_records
        )
        print("✅ Progress Inserted!")

# Executar a inserção de dados
insert_data()

# Fechar conexão com Neo4j
driver.close()
````

````python
%%writefile ai/prompts/context.md

### **📜 Context for AI to Generate Specific Questions on Student Activity in Neo4j**

#### **Introduction**
This database models student activity in an online academy, tracking **student purchases, course progress, and sales transactions**. The database is structured in **Neo4j**, a graph database, where relationships define the connections between students, courses, and their activities.

---

### **📌 Database Structure & Rules**
The system follows strict rules to ensure data consistency and accuracy:

#### **📍 Entities & Relationships**
- **Student (:Student)**
  - Represents a student enrolled in the academy.
  - Has attributes: `id`, `name`, `email`, `phone`.

- **Course (:Course)**
  - Represents an available course.
  - Has attributes: `name`, `url`.

- **Progress (:PROGRESS)**
  - Tracks a student's progress in a course.
  - Relationship: `(s:Student)-[:PROGRESS]->(c:Course)`.
  - Has attributes: `progress` (integer from 0 to 100).

- **Sales (:PURCHASED)**
  - Represents a student's purchase or refund of a course.
  - Relationship: `(s:Student)-[:PURCHASED]->(c:Course)`.
  - Has attributes: `status` ("paid" or "refunded"), `paymentMethod` ("pix" or "credit_card"), `paymentDate`, `amount`.

---

### **✅ Business Rules (Data Integrity Constraints)**
1. **A student can only have progress in a course they purchased.**
   - A `PROGRESS` relationship can only exist if there is a `PURCHASED` relationship where `status = "paid"`.

2. **A student can only have one progress entry per course.**
   - If a `PROGRESS` relationship already exists, the `progress` value is updated instead of creating a new relationship.

3. **A student can only purchase or refund a course once.**
   - A `PURCHASED` relationship can only exist once per `(Student, Course)`. If a new status is set, the existing record is updated.

---

### **📌 Cypher Queries Mapping the Rules**

#### **1️⃣ Verify if a Student Exists**
```cypher
MATCH (s:Student) RETURN s LIMIT 5;
```

#### **2️⃣ Verify Student's Course Progress (Must Have Purchased the Course)**
```cypher
MATCH (s:Student)-[:PURCHASED {status: "paid"}]->(c:Course)
OPTIONAL MATCH (s)-[p:PROGRESS]->(c)
RETURN s.name AS Student, c.name AS Course, p.progress AS Progress;
```

#### **3️⃣ Ensure a Student Can Only Buy or Refund a Course Once**
```cypher
MATCH (s:Student)-[p:PURCHASED]->(c:Course)
RETURN s.name, c.name, p.status, p.paymentMethod, p.paymentDate, p.amount;
```

#### **4️⃣ Find Students Who Have Progress Without Purchase (Invalid Case)**
```cypher
MATCH (s:Student)-[p:PROGRESS]->(c:Course)
WHERE NOT EXISTS {
    MATCH (s)-[:PURCHASED {status: "paid"}]->(c)
}
RETURN s.name AS Student, c.name AS Course, p.progress AS Progress;
```

#### **5️⃣ Find Students Who Bought but Never Started a Course**
```cypher
MATCH (s:Student)-[:PURCHASED {status: "paid"}]->(c:Course)
WHERE NOT (s)-[:PROGRESS]->(c)
RETURN s.name AS Student, c.name AS Course;
```

---

### **🚀 AI Question Examples Based on This Context**

#### ✅ **Valid Scenarios**
1. "Which students have purchased a course and started their progress?"
2. "What is the average progress of students who purchased 'Mastering Node.js Streams'?"
3. "List all students who completed a course (progress = 100%)."

#### ❌ **Edge Cases & Invalid Data Detection**
4. "Are there students with progress in a course they never bought?"
5. "Has any student purchased a course multiple times?"
6. "Which students purchased a course but never started it?"

---

### **🎯 Summary**
- **Neo4j is used to track student activity** in the academy.
- **Graph relationships enforce business rules** for purchases and progress.
- **Data integrity constraints prevent inconsistencies**.
- **AI-generated questions should respect these rules** and help detect invalid cases.
````

````python
%%writefile ai/prompts/responseTemplateFromJson.md

Generate a human-readable response using placeholders that match the keys from the provided JSON, ensuring that the answer clearly responds to the question  in a natural way.

### Rules:
- **Do not** generate SQL queries, JSON structures, or code snippets.
- Group results dynamically based on the first key in the data.
- Use placeholders **for each field** dynamically.
- **Do not include example JSON data in the output.**
- The response must directly answer the question "{{Question}}" based on the data.

### Question
{question}

### Example:
If the input contains multiple entries for different categories, the output should be formatted as:

"**{{GroupKey}}**:
{{EntriesList}}"

Where:
- `{{GroupKey}}` is the **first key in the dataset**.
- `{{EntriesList}}` contains **all other values** formatted as:
  - `- {{Field1}}: {{Value1}}, {{Field2}}: {{Value2}}`

Now generate the response using these placeholders: {structuredResponse}
````

````python
import os
import json
from langchain_community.graphs.neo4j_graph import Neo4jGraph
from langchain_ollama import ChatOllama, OllamaEmbeddings
from langchain_neo4j import Neo4jVector
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnableSequence
from aiofiles import open as aio_open

prompts_folder = 'ai/prompts'
prompts_files = {
    'nlpToCypher': f'{prompts_folder}/nlpToCypher.md',
    'responseTemplateFromJson': f'{prompts_folder}/responseTemplateFromJson.md',
    'context': f'{prompts_folder}/context.md',
}

# Load Neo4j Credentials
config = {
    'url': os.getenv('NEO4J_URI'),
    'username': "neo4j",
    'password': os.getenv('NEO4J_PASSWORD'),
    'indexName': "agent_index",
    'searchType': "vector",
    'textNodeProperties': ["question"],
    'nodeLabel': "Chunk",
}

# Initialize Models
coder_model = ChatOllama(
    temperature=0,
    maxRetries=2,
    model="deepseek-r1:14b",
    # baseURL=os.getenv('OLLAMA_BASE_URL'),
)

nlp_model = ChatOllama(
    temperature=0,
    maxRetries=2,
    model="deepseek-r1:14b",
    # baseURL=os.getenv('OLLAMA_BASE_URL'),
)

ollama_embeddings = OllamaEmbeddings(
    model="nomic-embed-text",
    # baseUrl=os.getenv('OLLAMA_BASE_URL'),
)

async def prompt(question, debug_log=lambda *args, **kwargs: None):
    # Initialize Neo4j Graph Connection
    graph = await Neo4jGraph.initialize(
        url=config['url'],
        username=config['username'],
        password=config['password'],
        enhancedSchema=False,
    )

    vector_index = await Neo4jVector.from_existing_graph(ollama_embeddings, config)

    # LangChain Pipeline
    chain = RunnableSequence.from_([
        retrieve_vector_search_results,
        generate_query_if_no_cached,
        validate_and_execute_query,
        generate_nlp_response,
        cache_result,
        parse_template_to_data,
    ])

    result = await chain.invoke({'question': question})
    debug_log("\n🎙️ Question:")
    debug_log("\n", question, "\n")
    debug_log(result.get('answer') or result.get('error'))

    await vector_index.close()
    await graph.close()

    return result

async def retrieve_vector_search_results(input):
    debug_log("🔍 Searching Neo4j vector store...")
    vector_results = await vector_index.similarity_search_with_score(input['question'], 1)
    results = vector_results[0] if vector_results else None
    score = results[1] if results else None

    if results and score > float(os.getenv('NEO4J_VECTOR_THRESHOLD', 0)):
        debug_log(f"✅ Vector match found! - score: {score}")
        return {
            **input,
            'cached': True,
            'answerTemplate': results[0]['metadata']['answerTemplate'],
            'query': results[0]['metadata']['query']
        }

    debug_log("⚠️ No vector match found, generating Cypher query...")
    return {
        **input,
        'cached': False,
    }

async def generate_query_if_no_cached(input):
    if input.get('cached'):
        return input  # Skip if we already have a cached answer

    schema = await graph.get_schema()
    async with aio_open(prompts_files['nlpToCypher'], 'r') as f:
        nlp_to_cypher_prompt = await f.read()
    async with aio_open(prompts_files['context'], 'r') as f:
        context = await f.read()
    query_prompt = ChatPromptTemplate.from_template(nlp_to_cypher_prompt)

    query_chain = query_prompt.pipe(coder_model).pipe(StringOutputParser())
    query = await query_chain.invoke({
        'question': input['question'],
        'schema': schema,
        'context': context
    })

    return {**input, 'query': query}

async def validate_and_execute_query(input):
    if input.get('cached'):
        db_results = await graph.query(input['query'])
        if not db_results:
            debug_log("⚠️ No meaningful results from Neo4j.")
            return {'error': "No results found."}

        return {**input, 'dbResults': db_results}

    debug_log("🤖 AI Generated Cypher Query:\n", input['query'])
    validation_result = await graph.query(f'EXPLAIN {input["query"]}')
    if not validation_result:
        debug_log("❌ Generated query is invalid:", input['query'])
        return {'error': "I couldn't generate a valid query."}

    db_results = await graph.query(input['query'])
    if not db_results:
        debug_log("⚠️ No meaningful results from Neo4j.")
        return {'error': "No results found."}

    return {**input, 'dbResults': db_results}

async def generate_nlp_response(input):
    if input.get('cached'):
        return input  # Skip if cached
    if input.get('error'):
        return input  # Handle errors
    async with aio_open(prompts_files['responseTemplateFromJson'], 'r') as f:
        response_template_prompt = await f.read()
    response_prompt = ChatPromptTemplate.from_template(response_template_prompt)

    response_chain = response_prompt.pipe(nlp_model).pipe(StringOutputParser())

    # Ensure structuredResponse is formatted as a string
    ai_response = await response_chain.invoke({
        'question': input['question'],
        'structuredResponse': json.dumps(input['dbResults'][0])  # Ensure JSON data is properly formatted
    })

    return {**input, 'answerTemplate': ai_response}

def parse_template_to_data(input):
    if input.get('error'):
        return input
    if not input.get('dbResults'):
        return {
            **input,
            'answer': "I'm sorry, but I couldn't find any relevant information."
        }

    # Ensure we have a valid template
    template = input.get('answerTemplate', "**Results:**\n{Results}")
    # Extract placeholders from the template
    placeholders = re.findall(r'{(.*?)}', template)

    # Extract the static part (before the first placeholder)
    static_header, dynamic_template = template.split("\n\n", 1) if "\n\n" in template else (template, '')

    # Process each entry and replace placeholders
    formatted_entries = []
    for entry in input['dbResults']:
        formatted_entry = dynamic_template or template  # Use the part after the first newline

        for placeholder in placeholders:
            key = placeholder.strip('{}')  # Remove { }
            value = entry.get(key, "")
            formatted_entry = formatted_entry.replace(f'{{{key}}}', str(value))

        formatted_entries.append(formatted_entry)

    return {
        **input,
        'answer': f"{static_header}\n\n" + "\n\n".join(formatted_entries)
    }
````

````python
from flask import Flask, request, jsonify
import asyncio

app = Flask(__name__)

# Função simulada para prompt (substitua pela sua implementação real)
async def prompt(question: str, log_function) -> dict:
    log_function(f"Prompting AI for question: {question}")
    await asyncio.sleep(1)  # Simula uma operação assíncrona
    return {"answer": f"Resposta simulada para: {question}", "error": None}

# Função de log de depuração
def debug_log(*args):
    print(*args)

@app.route('/v1/chat', methods=['POST'])
def chat():
    try:
        data = request.json
        log_function = debug_log
        log_function("🔹 Received AI Prompt:", data['prompt'])

        # Executa a função assíncrona prompt dentro de um loop de eventos
        ai_response = asyncio.run(prompt(data['prompt'], log_function))

        return jsonify(ai_response)

    except Exception as error:
        print("❌ AI Backend Error:", str(error))
        return jsonify({"message": "Internal Server Error"}), 500

if __name__ == '__main__':
    app.run(port=3002)  # O Google Colab geralmente redireciona o tráfego para a porta 8080
````

````python
#data/seed.js
from neo4j import GraphDatabase
from faker import Faker
import json
import os

# Configuração do Neo4j
NEO4J_URI = os.getenv("NEO4J_URI", "bolt://localhost:7687")
NEO4J_USER = os.getenv("NEO4J_USER", "neo4j")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD", "password")

# Criar driver do Neo4j
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USER, NEO4J_PASSWORD))

# Instanciar Faker para gerar dados fictícios
faker = Faker()

# Carregar cursos a partir de um arquivo JSON
with open("./data/courses.json", "r", encoding="utf-8") as file:
    courses = json.load(file)

# Criar alunos fictícios
students = [
    {
        "id": faker.uuid4(),
        "name": faker.name(),
        "email": faker.email(),
        "phone": faker.phone_number(),
    }
    for _ in range(20)
]

# Criar registros de vendas
sales_records = [
    {
        "studentId": student["id"],
        "courseId": faker.random_element(courses)["name"],
        "status": faker.random_element(["paid", "refunded"]),
        "paymentMethod": faker.random_element(["pix", "credit_card"]),
        "paymentDate": faker.past_date().isoformat(),
        "amount": round(faker.pyfloat(min_value=0, max_value=2000, right_digits=2), 2),
    }
    for student in students
    for _ in range(faker.random_int(min=1, max=10))
]

# Criar registros de progresso para vendas pagas
progress_records = [
    {
        "studentId": sale["studentId"],
        "courseId": sale["courseId"],
        "progress": faker.random_int(min=0, max=100),
    }
    for sale in sales_records
    if sale["status"] == "paid"
]

# Função para inserir dados no Neo4j
def insert_data():
    with driver.session() as session:
        # Limpar banco de dados (opcional)
        # session.run("MATCH (n) DETACH DELETE n")
        # print("🧹 Database cleared!")

        session.run(
            """
            UNWIND $batch AS row
            MERGE (c:Course {name: row.name})
            ON CREATE SET c.url = row.url
            """,
            batch=courses
        )
        print("✅ Courses Inserted!")

        session.run(
            """
            UNWIND $batch AS row
            MERGE (s:Student {id: row.id})
            ON CREATE SET s.name = row.name, s.email = row.email, s.phone = row.phone
            """,
            batch=students
        )
        print("✅ Students Inserted!")

        session.run(
            """
            UNWIND $batch AS row
            MATCH (s:Student {id: row.studentId}), (c:Course {name: row.courseId})
            MERGE (s)-[p:PURCHASED]->(c)
            ON CREATE SET p.status = row.status, p.paymentMethod = row.paymentMethod,
                          p.paymentDate = row.paymentDate, p.amount = row.amount
            ON MATCH SET p.status = row.status, p.paymentMethod = row.paymentMethod,
                         p.paymentDate = row.paymentDate, p.amount = row.amount
            """,
            batch=sales_records
        )
        print("✅ Sales Inserted!")

        session.run(
            """
            UNWIND $batch AS row
            MATCH (s:Student {id: row.studentId})-[:PURCHASED {status: "paid"}]->(c:Course {name: row.courseId})
            MERGE (s)-[p:PROGRESS]->(c)
            ON CREATE SET p.progress = row.progress
            ON MATCH SET p.progress = row.progress
            """,
            batch=progress_records
        )
        print("✅ Progress Inserted!")

# Executar a inserção de dados
insert_data()

# Fechar conexão com Neo4j
driver.close()
````

````python
from langchain_community.graphs.neo4j_graph import Neo4jGraph
from langchain_ollama import ChatOllama, OllamaEmbeddings
from langchain_community.vectorstores.neo4j_vector import Neo4jVectorStore
import os

# ✅ Load Neo4j credentials from environment variables
config = {
    "url": userdata.get('NEO4J_URI'),
    "username": "neo4j",
    "password": userdata.get('NEO4J_PASSWORD'),
    "text_node_properties": ["text"],
    "index_name": "sim_example_index",
    "keyword_index_name": "sim_example_keywords",
}

# ✅ Initialize Ollama Embeddings Model
ollama_embeddings = OllamaEmbeddings(model="nomic-embed-text", base_url=os.getenv("OPENAI_BASE_URL"))

# ✅ Create Neo4j Vector Store
neo4j_vector_index = Neo4jVectorStore.from_existing_graph(ollama_embeddings, config)

# ✅ Documents to Store in Neo4j
documents = [
    {"page_content": "the author who commented most is Erick", "metadata": {}},
    {"page_content": "the less active author is Ana", "metadata": {}},
    {"page_content": "the post abc is the one who received less comments", "metadata": {}},
    {"page_content": "the post ewacademy is the one who received more comments", "metadata": {}},
]

# ✅ Function to Check and Add Documents
def add_document_if_not_exists(doc):
    search_results = neo4j_vector_index.similarity_search_with_score(doc["page_content"], 1)
    if search_results:
        score, item = search_results[0]
        print("🔍 Search Results:", search_results, score)
        if score > 0.9 and item.page_content == f"\ntext: {doc['page_content']}":
            print(f"🚫 Skipping duplicate: \"{doc['page_content']}\"")
            return
    print(f"✅ Adding new document: \"{doc['page_content']}\"")
    neo4j_vector_index.add_documents([doc])

# ✅ Iterate Over Documents and Add Only If Not Exists
for doc in documents:
    add_document_if_not_exists(doc)

# ✅ Function to Ask a Question
def make_a_question(question):
    results = neo4j_vector_index.similarity_search_with_score(question, 1)
    if results:
        print("🔍 Search Results:", question, results[0][1], results[0][0])

make_a_question("which one is the most popular post?")
make_a_question("which one is the less popular post?")
make_a_question("which one is top post?")
make_a_question("which one is worst post?")

# ✅ Close Neo4j Connection
neo4j_vector_index.close()
````
