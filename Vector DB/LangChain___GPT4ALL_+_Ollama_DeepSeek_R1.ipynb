#**Escolhendo o modelo**

Há uma ampla variedade de modelos de LLM (Large Language Models) disponíveis para uso, tanto por meio de APIs quanto em execução local. A escolha do modelo mais adequado depende exclusivamente da tarefa à qual ele será submetido, pois cada um possui suas próprias limitações. Alguns modelos foram treinados especificamente para desempenhar determinadas funções.

Um exemplo é o **Qwen2-1.5B-Instruct**, um modelo desenvolvido para realizar a extração de texto de documentos de forma eficiente. Ele é amplamente utilizado em RAG (*Retrieval-Augmented Generation*), uma técnica de inteligência artificial (IA) que aprimora as respostas de modelos de linguagem grande (LLM) ao incorporar um sistema de recuperação de informações. Esse sistema fornece dados externos ao LLM, permitindo-lhe gerar respostas mais precisas e contextualizadas.

Por essa razão, o **Qwen2-1.5B-Instruct** é um modelo leve; no entanto, não possui um contexto extenso. Em outras palavras, ele não foi treinado para armazenar uma grande quantidade de informações, mas sim para interpretar e compreender os dados fornecidos pelo usuário, garantindo uma extração de conteúdo eficiente e precisa.

Outro exemplo é o **Rift Coder**, que não é um modelo voltado para interações em bate-papo. Seu propósito é auxiliar na edição contextual de código, sendo utilizado principalmente por agentes para a geração de código.

Por outro lado, há modelos como o **Llama 3 8B Instruct**, que foi treinado para interagir com o usuário em conversas e possui um vasto conjunto de dados, permitindo-lhe compreender diferentes contextos. No entanto, pode não ser a melhor escolha para a geração de código, uma vez que sua especialização é a comunicação e a assistência por meio de diálogo.     

#**GPT4ALL**

O GPT4All é um ecossistema de código aberto utilizado para integrar LLMs em um único aplicativo, sem a necessidade de pagar por uma assinatura de plataforma ou por hardware específico. Ele foi criado pela Nomic AI, uma empresa especializada em cartografia de informações que busca melhorar o acesso aos recursos de inteligência artificial.

O GPT4All foi projetado para ser executado em computadores com hardware modesto, sem exigir conexão com a Internet ou mesmo uma GPU. Isso é possível porque a maioria dos modelos fornecidos pelo GPT4All foi quantizada, reduzindo seu tamanho para apenas alguns gigabytes, o que permite sua execução com apenas 4 a 16 GB de RAM.

Além disso, o GPT4All é focado na transparência e na privacidade dos dados. Suas informações permanecerão armazenadas exclusivamente no seu hardware local, a menos que você utilize uma chave de API para acessar provedores de modelos externos. Dessa forma, é possível utilizar LLMs disponibilizados por uma ampla variedade de provedores, como OpenAI, Anthropic, Cohere e Google AI. No entanto, ao optar por esse recurso, perde-se a privacidade garantida pelo uso exclusivamente local.

Outro ponto relevante é que o GPT4All é gratuito e de código aberto. Vários dos LLMs fornecidos são licenciados sob a GPL-2, permitindo que qualquer pessoa ajuste e integre seus próprios modelos para uso comercial sem a necessidade de pagar pelo licenciamento.

Através do código abaixo, é possível visualizar os modelos disponíveis no GPT4All. Na descrição dos LLMs, é possível verificar suas características específicas, tais como:

* O tipo de modelo, que pode ser voltado para chat, geração de código ou embedding;
* A velocidade de processamento, categorizada como rápida, muito rápida ou lenta;
* O contexto geral do modelo, indicando a quantidade de dados utilizada em seu treinamento;
* O tamanho do modelo, expresso em parâmetros, como 1B, 3B, 8B ou 14B;
* O tipo de prompt suportado, como ChatML, Phi-3 ou Llama 3;
* A compatibilidade com múltiplos idiomas (suporte multilíngue).

```python
import json
import requests
from bs4 import BeautifulSoup

# URL do arquivo JSON
url = "https://raw.githubusercontent.com/nomic-ai/gpt4all/main/gpt4all-chat/metadata/models3.json"

# Fazer o download do conteúdo do JSON
response = requests.get(url)
models_data = response.json()

# Exibir informações sobre cada modelo
for model in models_data:
    print(f"Nome: {model['name']}")
    print(f"Parâmetros: {model['parameters']}")
    print(f"Tipo: {model['type']}")

    # Converter HTML da descrição para texto formatado
    soup = BeautifulSoup(model['description'], "html.parser")
    description_items = [li.get_text() for li in soup.find_all("li")]

    print("Descrição:")
    for item in description_items:
        print(f"  - {item}")

    print(f"URL: {model['url']}")
    print("-" * 40)
```

#**Baixando modelos**

Existe muitas formas de você baixar os modelos LLM para utilizar em seu projeto. Aqui vou mostrar três maneiras de você fazer isso:

* através da `huggingface_hub`, biblioteca desenvolvidade pela Hugging Face.  Ela serve como interface para interagir com a plataforma Hugging Face Hub, permitindo baixar modelos, datasets, tokenizadores e outros recursos de IA. A sua utilização é necessária quando pretende-se utilizar *modelos privados* (modelos que precisam de autorização para serem utilizados) do Hugging Face.

* através do `wget`, utilitário de linha de comando usado para baixar arquivos da internet. Ele suporta os protocolos HTTP, HTTPS e FTP e é muito utilizado para baixar páginas da web, arquivos ou até mesmo espelhar sites inteiros.

* através do `Aria`, gerenciador de downloads de linha de comando: rápido, leve e multi-protocolo (pode baixar arquivos via HTTP/HTTPS, FTP, SFTP, BitTorrent e Metalink). Ele é uma alternativa mais avançada ao wget, pois permite download de multiplas fontes, múltiplas conexões paralelas, retomada de downloads interrompidos e suporte a vários protocolos.

Dentre esses três métodos, o gerenciador de downloads `Aria` se sobressai aos demais, pois consegue fragmentar o download por meio de conexões paralelas, o que contribui para que o processo de download seja realizado de forma eficiente e mais rápida. Além disso, o `Aria` é compatível com uma ampla variedade de protocolos, facilitando sua utilização nos mais diversos projetos.

###Métodos de utilização do LangChain

Existem várias configurações e parâmetros que podem ser utilizados para melhorar a apresentação dos resultados obtidos por meio de uma LLM. Aqui, abordaremos os conceitos de *callbacks* e *output parsers*:

* `callbacks`: O Langchain fornece fornece um sistema abrangente para  [eventos de *callbacks* (retorno de chamada)](https://python.langchain.com/docs/concepts/callbacks/), permitindo a conexão com diversos estágios do desenvolvimento de aplicativos, agentes e chatbots baseados em LLM. Isso é útil para registro, monitoramento, streaming e outras tarefas.

* `output_parsers`: Os *output parsers* são responsáveis por processar a saída de uma LLM e convertê-la em um formato mais estruturado. Lembre-se de que LLMs podem apresentar inconsistências! Para garantir a correta formatação da saída, é necessário utilizar um LLM com capacidade suficiente para gerar um formato de saída bem estruturado.

  Existem diversos formatos de saída, como JSON, XML e YAML.
  
  O `JsonOutputParser`, por exemplo, é uma opção interna que permite solicitar e posteriormente analisar a saída em JSON. Ele é extremamente útil para converter o resultado em JSON, pois, mesmo que a saída de uma LLM tenha a mesma estrutura de um JSON, ela pode ser interpretada de forma diferente. Isso ocorre porque a saída de uma LLM é, na verdade, uma string (texto), e não um objeto JSON nativo.

Com utilização de `callbacks`:

```python
from typing import Any, Dict, List

from langchain_community.llms import GPT4All
from langchain_core.callbacks import BaseCallbackHandler
from langchain_core.messages import BaseMessage
from langchain_core.outputs import LLMResult
# from langchain_core.prompts import ChatPromptTemplate

# class MyCustomHandler(BaseCallbackHandler):
#     def on_llm_new_token(self, token: str, **kwargs) -> None:
#         global count
#         if count < 10:
#             print(f"Token: {token}")
#             count += 1

# callbacks = [MyCustomHandler()]

class LoggingHandler(BaseCallbackHandler):
    def on_chat_model_start(
        self, serialized: Dict[str, Any], messages: List[List[BaseMessage]], **kwargs
    ) -> None:
        print("Chat model started")

    def on_llm_end(self, response: LLMResult, **kwargs) -> None:
        print(f"Chat model ended, response: {response}")

    def on_chain_start(
        self, serialized: Dict[str, Any], inputs: Dict[str, Any], **kwargs
    ) -> None:
        print(f"Chain {serialized.get('name')} started")

    def on_chain_end(self, outputs: Dict[str, Any], **kwargs) -> None:
        print(f"Chain ended, outputs: {outputs}")

callbacks = [LoggingHandler()]

local_path = (
    "./models/qwen2-1_5b-instruct-q4_0.gguf"
)

llm = GPT4All(model=local_path,  max_tokens=1000, callbacks=callbacks, streaming=False)
chain = prompt | llm

# chain = prompt | llm
# chain_with_callbacks = chain.with_config(callbacks=callbacks)

question = "What NFL team won the Super Bowl in the year Justin Bieber was born?"

# Streamed tokens will be logged/aggregated via the passed callback
chain.invoke({"question": question})
```

Com utilização de `output_parsers`:

```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_community.llms import GPT4All
from pydantic import BaseModel, Field

local_path = (
    "./models/qwen2-1_5b-instruct-q4_0.gguf"
)

llm = GPT4All(model=local_path, max_tokens=2000, temp=0.7, #top_k=40, top_p=0.9,
                      repeat_penalty=1.1, streaming=True)


# Define your desired data structure.
class Joke(BaseModel):
    setup: str = Field(description="question to set up a joke")
    punchline: str = Field(description="answer to resolve the joke")


# And a query intented to prompt a language model to populate the data structure.
joke_query = "Tell me a joke."

# Set up a parser + inject instructions into the prompt template.
parser = JsonOutputParser(pydantic_object=Joke)

prompt = PromptTemplate(
    template="Answer the user query.\n{format_instructions}\n{query}\n",
    input_variables=["query"],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

chain = prompt | llm | parser

chain.invoke({"query": joke_query})
```

O OLLAMA é um ecossistema de código aberto utilizado para rodar localmente LLMs em um único aplicativo.

## Executando o Ollama
---

Para utilizar o Ollama, é necessário que ele seja executado como um serviço em segundo plano, paralelamente aos seus scripts.

Como o Jupyter Notebooks foi projetado para executar blocos de código em sequência, isso dificulta a execução simultânea de dois blocos. Como solução alternativa, criaremos um serviço utilizando o módulo subprocess em Python, garantindo que nenhum bloco de código seja bloqueado durante a execução.

O serviço pode ser iniciado com o seguinte comando: `ollama serve`.

Além disso, a função `time.sleep(5)` adiciona um pequeno atraso para garantir que o serviço Ollama esteja completamente ativo antes de iniciar o download do modelo.

```python
import threading
import subprocess
import time

def run_ollama_serve():
  subprocess.Popen(["ollama", "serve"])

thread = threading.Thread(target=run_ollama_serve)
thread.start()
time.sleep(5)
```
